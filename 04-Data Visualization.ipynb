### 1-1 寫一個文字檔案來儲存

先介紹開啟檔案的參數：  

<pre>
    r - 唯讀（read）  - 只能讀取檔案
    w - 寫入（write） - 可以寫入檔案，會覆蓋原有內容
    a - 附加（append）- 可以寫入檔案，不會覆蓋原有內容
</pre>

+ 試著修改寫入文件檔案的內容或是檔名

```python
#開啟檔案或是建立新檔案
file = open("newfile.txt","w")

#寫入文字
file.write("r - 唯讀（read）  - 只能讀取檔案\n") 
file.write("w - 寫入（write） - 可以寫入檔案，會覆蓋原有內容\n") 
file.write("a - 附加（append）- 可以寫入檔案，不會覆蓋原有內容\n") 

#關閉檔案
file.close() 
```

### 補充：寫入檔案也有很多其他有趣的寫法

#### 利用print

```python
#開啟檔案或是建立新檔案
file = open("newfile.txt","a")

#用print寫入文字
print("我這次用print寫！\n",file=file)

#關閉檔案
file.close()
```

#### 使用writelines一次寫入好多行

```python
#開啟檔案或是建立新檔案
file = open("newfile.txt","a")

#建立一個lines清單放所有想寫入的內容
lines = ["我這次\n","使用\n","writelines\n","寫！\n"]

#使用writelines一次將lines清單中所有內容寫入
file.writelines(lines)

#關閉檔案
file.close()
```

#### 優雅的使用with寫入檔案，不怕忘記用close()關閉檔案

```python
#開啟檔案或是建立新檔案後寫入
with open("newfile.txt","a") as file:
    file.write("使用with寫入！")
```

### 1-2 讀取文字檔案

使用讀取檔案功能 - r，還有read()來協助讀取文字檔案。

```python
#用唯讀模式開啟檔案
file = open("newfile.txt", "r") 

#印出讀取的文字檔案內容
print(file.read()) 

#關閉檔案
file.close()
```

==============

在read()括號之間加數字代表你要讀取文字檔案中的**前幾個字元**

```python
#用唯讀模式開啟檔案
file = open("newfile.txt", "r") 

#讀前十二個字元
print(file.read(12)) 

#關閉檔案
file.close()
```

==============  

readline()可以讀取**文字檔案第一行**

```python
#用唯讀模式開啟檔案
file = open("newfile.txt", "r") 

#讀取文字檔案第一行
print(file.readline()) 

#關閉檔案
file.close()
```

==============  

readlines()可以獲得文字檔案**以行為單位的清單**

```python
#用唯讀模式開啟檔案
file = open("newfile.txt", "r") 

#獲得文字檔案以行為單位的清單
print(file.readlines()) 

#關閉檔案
file.close()
```

==============  
當一個文字檔案有很**多行**，且你需要**按順序處理每一行**時，可以使用**for迴圈**來協助檔案的讀取！

```python
#用唯讀模式開啟檔案
file = open("newfile.txt", "r") 

#計算行數用的變數
i = 1

#使用for迴圈按順序處理每一行
for line in file:
    print("第"+str(i)+"行："+line)
    i+=1  #行數變數加一
    
#關閉檔案
file.close()
```

設定線的顏色跟樣式

1.顏色

2.線的樣式
<pre>

```python
##補充：不同顏色的設定方法一：RGB(A)

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(1,figsize=(10,5))

###############顏色設定順序：紅、綠、藍、透明度（這一項可有可無）
plt.plot(x,y1,color=(120/255,120/255,255/255,0.8))
plt.plot(x,y2,color=(255/255,150/255,255/255,0.5))
#########################################################

plt.show()
```

```python
##補充：不同顏色的設定方法二:十六進位色碼

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(1,figsize=(10,5))

##################################
plt.plot(x,y1,color='#3AAD1F')
plt.plot(x,y2,color='#AF87F1')
##################################

plt.show()
```

### 2-3 直方圖

補充資料(直方圖中寬度等細節設定): https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.hist.html

當資料有連續性，且你想要看出資料分布的狀況時，適合使用直方圖

範例中使用整班學生的成績為資料，成績一個一個打太累了，我們使用numpy函式庫中的rabdom.randint()，隨機產生100組0-100的值，再將這些值做簡單處理



```python
scores = np.random.randint(101,size=100)
print(scores)
```

```python
#直方圖
plt.hist(scores,width=8)
plt.show()
```


+ 試試看用append增加學生的成績，會有什麼變化呢？

```python
#將numpy array變成list
scores_list=scores.tolist()

#新增一個成績
scores_list.append(70)

#查看70有沒有被成功新增
print(scores_list)

#將list變成numpy array
scores=np.array(scores_list)

#查看兩個變數的類別
print(type(scores_list))
print(type(scores))

#顯示直方圖
plt.hist(scores)
plt.show()
```

### 2-5 圓餅圖

當你想要看資料的以百分比呈現時使用。

參考資料:
https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.pie.html

+ 試試看修改每個動物類別的喜愛人數吧!


```python
# 設定顯示圖框架大小
plt.figure(figsize=(10,10)) 

# 類別標籤
labels = ["mouse","dog","cat","rabbit","bird"] 

# 設定各類別數值
size = [100,24,34,16,55]                        

# 設定圓餅圖細節
plt.pie(size,                           # 數值
        labels = labels,                # 標籤
        autopct = "%0.1f%%",            # 將數值取到小數點一位
        textprops = {"fontsize" : 12},  # 文字大小
       ) 

 
# 設定標題及其文字大小
plt.title("Pie chart of favorite animal", {"fontsize" : 18})

# 設定圖例放在最好的位置
plt.legend(loc = "best")   

plt.show()
```

### 3-2. 匯入地形資料

```python
counties_shp = gpd.read_file('COUNTY_MOI_1081121.shp',encoding='utf-8')
town_shp = gpd.read_file('TOWN_MOI_1081121.shp',encoding='utf-8')
```

### 3-3. 繪製台灣地圖

```python
fig = counties_shp.plot(column='COUNTYNAME')
ax = fig.axis((119.2,122.5,21.5,25.5))
```

### 3-5. 繪製台灣大圖

```python
fig, ax = plt.subplots(1, figsize=(20, 20))
ax = counties_shp.plot(ax=ax)
ax.set_xlim((119.2,122.5))
ax.set_ylim((21.5,25.5))

#加上標題
fig.suptitle('Taiwan counties')
#顯示圖表
plt.show()
```

### 3-6. 刪除縣市、檢視縣市細節資訊

```python
for index, row in counties_shp.iterrows():
    if row['COUNTYNAME'] == '連江縣' or row['COUNTYNAME'] == '金門縣' or row['COUNTYNAME'] == '澎湖縣'  :
        counties_shp.drop(index, inplace=True)
#顯示表格
counties_shp
```

### 3-7. 更詳盡的台灣地圖

### 3-8. 更詳盡的單一縣市地圖 

```python
fig, ax = plt.subplots(1, figsize=(10, 10))
taichung_town_shp=town_shp[town_shp['COUNTYNAME']=='新北市']
ax = taichung_town_shp.plot(ax=ax,cmap='RdBu')

fig.suptitle('New Taipei map',fontsize=35)
plt.show()
```
